```python
import pandas as pd
import requests
import json
import datetime
```

```python
df_redefine = pd.read_csv('../raw_data/ariel_project_data_1Kdays.csv')
```

```python
df_redefine['other_contracts'] = df_redefine['other_contracts'].apply(lambda x: x.replace("'",'"'))
df_redefine['other_contracts'] = df_redefine['other_contracts'].apply(lambda x: json.loads(x))
```

```python

def get_token_date(address):
    print(address)
    api_key = 'c1cafc9218c5575879f12eedf6822e'
    url = 'https://redefine-api.net/external/v1/riscore?address='
    # Make the GET request
    headers = {
    "X-API-Key": api_key
    }
    response = requests.get(url+address,headers=headers)
    data = response.json()
    time=data['result']['riscore_metrics']['creation_info']['creation_time']
    # Convert the Unix time to a datetime object
    datetime_obj = datetime.datetime.fromtimestamp(time)
    format_string = '%Y-%m-%d'
    formatted_datetime = datetime_obj.strftime(format_string)
    # Print the datetime object
    return formatted_datetime

df_redefine['token_creation_date'] = df_redefine['token_address'].apply(lambda x: get_token_date(x))
df_redefine

```
